# Python

## Basic Data Types

**Numbers:** int & floats

```python
x = 10

print(type(x))
print(x)
print(x + 1)
print(x - 1)

x += 1
print(x)  # print "11"

x *= 2
print(x)  # print "22"
```

**Booleans:** Python implements all of the usual operators for Boolean logic, but uses english words rather than symbols

```python
t = True
f = False

print(type(t))
print(type(f))

print(t and f)
print(t or f)
print(not t)
print( t != f)
```

**Strings:** Python has great support for strings

```python
hello = 'hello'
world = 'world'

print(hello)
print(len(hello))

hw = hello + ' ' + world
print(hw)
```

```python
s = "hello"

print(s.capitalize())
print(s.upper())
print(s.replace('l', '(ell)'))
print('  world '.strip())
```

## Containers

**Lists:** A list is the Python equivalent of an array, but is resizeable and can contain elements fo different types

List is mutable

```python
xs = [3, 1, 2]
print(xs, xs[2]) # indexing

print(xs[-1]) # indexing

xs[2] = 'foo'
print(xs)

xs.append('bar')
print(xs)

x = xs.pop() #remove & return the last element
print(x, xs)
```

**Slicing:** In addition to accessing list elements one at a time, Python provides concise syntax to access sublists; this is known as slicing

```python
nums = list(range(5))

print(nums)
print(nums[2:4])
print(nums[2:])
print(nums[:2])
print(nums[:])
print(nums[:-1])
nums[2:4] = [8, 9]
print(nums)
```

**Loops:** You can loop over the elements of a list

```python
animals = ['cat', 'dog', 'monkey']
for animal in animals:
    print(animal)
```

If you want access to the index of each element within the body of a loop, use the built-in ``enumerate`` function:

```python
animals = ['cat', 'dog', 'monkey']

for idx, animal in enumerate(animals):
    print(f"#{idx}: {animal}")
```

**List Comprehension:** When programming, frequently we want to transform one type of data into another. 

```python
nums = [0, 1, 2, 3, 4]
squares = []
for x in nums:
    squares.append(x ** 2)
print(squares)   # Prints [0, 1, 4, 9, 16]

# using list comprehension
nums = [0, 1, 2, 3, 4]
squares = [x**2 for x in nums]
print(squares)
```
